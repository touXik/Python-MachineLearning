#### dataset seaborn  https://github.com/mwaskom/seaborn-data

```python
titanic =  sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.drop(['alone','alive','embark_town','adult_male','adult_male','who'],axis=1,inplace=True)
titanic.dropna(axis=0,inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

### pour les  Categories 

```python
sns.catplot(x='pclass',y='age',data=titanic,hue='sex')
```

```python
sns.boxplot(x='pclass',y='age',data=titanic,hue='sex')
```

### pour les  Distribution 

```python
sns.distplot(titanic['fare'])
```

```python
sns.jointplot('age','fare',data=titanic,kind="hex")
```

### Pour les Matrices

```python
sns.heatmap(titanic.corr())
```
