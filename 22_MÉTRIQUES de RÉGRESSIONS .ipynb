```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.metrics import * 
```

## mae mse 

```python
y = np.array([1,2])
y_pred = np.array([5,2])

print('MAE ',mean_absolute_error(y,y_pred))
print('MSE' ,mean_squared_error(y,y_pred))
print('RMSE' ,np.sqrt(mean_squared_error(y,y_pred)))
```

## median absolut error

```python
y = np.array([1,2,2,3,5,2])
y_pred = np.array([5,2,2,5,7,1000])

print('MAE : ',mean_absolute_error(y,y_pred))
print('MSE : ' ,mean_squared_error(y,y_pred))
print('RMSE :' ,np.sqrt(mean_squared_error(y,y_pred)))
print('median abs err : ',median_absolute_error(y,y_pred))
```

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
```

```python
boston = load_boston()
X =boston.data
y=boston.target

model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)

plt.scatter(X[:,5],y,label='y')
plt.scatter(X[:,5],y_pred,alpha=0.8,label='y_pred')
plt.legend()
```

```python
err_hist = np.abs(y-y_pred)
plt.hist(err_hist,bins=50)
plt.show()
```

## coffecion de determination R carre

```python
model.score(X,y)
```
