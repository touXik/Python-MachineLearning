### la fonction Resample 

```python
bitcoin['2019']['Close'].resample('M').plot()
plt.show()
```

```python
bitcoin['2019']['Close'].resample('M').std().plot()
plt.show()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2019','Close'].plot(label='anne 2019')
bitcoin.loc['2019','Close'].resample('M').mean().plot(label='Moyenne par mois',lw=3,ls=':',alpha=0.8)
bitcoin.loc['2019','Close'].resample('W').mean().plot(label='Moyenne par semaine',lw=2,ls='--',alpha=0.8)
plt.legend()
plt.show()
```

### fonction resample().agg() ===> aggregate 

```python
bitcoin['2019']['Close'].resample('W').agg(['mean','std','min','max']).head()
```

```python
m=bitcoin['2019']['Close'].resample('W').agg(['mean','std','min','max'])


plt.figure(figsize=(12,8))
m['mean']['2019'].plot(label='moyenne par semain')
plt.fill_between(m.index,m['min'],m['max'],alpha=0.2,label='min-max par semain')
```

### Moving AVERAGE 

### Rolling

```python
bitcoin.loc['2019','Close'].rolling(window=7).mean().plot()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2019-09','Close'].plot()
# bitcoin.loc['2019-09','Close'].rolling(window=7).mean().plot(ls=':',lw=3,label='moving average')
# bitcoin.loc['2019-09','Close'].rolling(window=7,center=True).mean().plot(ls=':',lw=3)
for i in np.arange(0.2,1,0.2):
    bitcoin.loc['2019-09','Close'].ewm(alpha=i).mean().plot(label='ewm {}'.format(i),ls=':',lw=3)
plt.legend()
plt.show()
```

# Assembler les data set ensemble 

## comparer les bitcoin avec les Etheruem 

```python
etheruem =pd.read_csv('data_set/ETH-EUR.csv',index_col='Date',parse_dates=True)
```

```python
etheruem['2019']['Close'].plot()
```

```python
btc_eth=pd.merge(bitcoin,etheruem,on='Date',how='inner',suffixes=('_btc','_eth'))
```

```python
btc_eth[['Close_btc','Close_eth']].plot(subplots=True,figsize=(12,8))
```

```python
btc_eth[['Close_btc','Close_eth']].corr()
```

```python
correlation=btc_eth[['Close_btc','Close_eth']].corr()
```

```python
import seaborn as sns
sns.heatmap(correlation)
```

# EXO TRading Strategy

```python
bitcoin.loc['2019','Close'].rolling(window=7).max().plot(label='max')
bitcoin.loc['2019','Close'].rolling(window=7).min().plot(label='min')
plt.legend()
```

```python
bitcoin.loc['2019-09-09':'2019-10-04','Close'].rolling(window=7).max().plot(label='max')
bitcoin.loc['2019-09-09':'2019-10-04','Close'].rolling(window=7).min().plot(label='min')
plt.legend()
```
