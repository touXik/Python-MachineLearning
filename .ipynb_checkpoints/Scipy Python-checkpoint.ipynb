# scipy 

### 1) interpolation

```python
from scipy.interpolate import interp1d
```

```python
x=np.linspace(0,10,10)
y=x**2
plt.scatter(x,y)
```

```python
f = interp1d(x,y,kind='cubic')

new_x=np.linspace(0,10,50)
result=f(new_x)

plt.scatter(x,y)
plt.plot(new_x,result,c='r')
```

### 2) optimize

```python
from scipy import optimize
```

```python
x=np.linspace(0,2,100)
y= 1/3* x**3 - 3/5 * x**2 + 2*x + np.random.randn(x.shape[0])/20
plt.scatter(x,y)
```

```python
def f(x, a,b,c,d):
    return  a*x**3+ b* x**2 + c + d
```

```python
param, param_cov = optimize.curve_fit(f,x,y)
plt.scatter(x,y)
plt.plot(x,f(x,param[0],param[1],param[2],param[3]),c='g',lw=3)
```

```python
def f(x):
    return x**2+15*np.sin(x)
```

```python
x = np.linspace(-10,10,100)
plt.plot(x,f(x))
```

```python
x0=-5
result=optimize.minimize(f,x0=x0).x
plt.plot(x,f(x))
plt.scatter(result,f(result),s=100,c='r')
plt.scatter(x0,f(x0),s=100,marker='+',c="g")
plt.show
```

#### Transformation de fourier
analyse les Frequence presentes dans un signal periodique 
en resultat : un spectre 

```python
x = np.linspace(0,30,1000)
y = 3*np.sin(x)+2*np.sin(5*x)+np.sin(10*x)+np.random.randn(x.shape[0])
plt.plot(x,y)
```

```python
from scipy import fftpack
```

```python
fourier =fftpack.fft(y)
power =np.abs(fourier)
frequences = fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences),power)
```

```python
fourier[power<400]=0
```

```python
plt.plot(np.abs(frequences),np.abs(fourier))
```

```python
filterd_signal=fftpack.ifft(fourier)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x,y,lw=0.5)
plt.plot(x,filterd_signal,lw=3)
plt.show
```
